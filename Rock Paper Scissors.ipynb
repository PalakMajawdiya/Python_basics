Rock, Paper, Scissors is played as a means to help make a choice or put an end to a disagreement. The idea is that both players have an equal chance of winning, making the game random but fair.

After you’ve both played an object, look to see who’s won. Each object wins against one shape and loses to another. For instance, rock “crushes” scissors but is “covered” by paper, paper “covers” rock but is “cut” by scissors, and scissors is “crushed” by rock but “cuts” paper. The player who picks the stronger of the two objects is the winner.
If both players throw the same object, it’s a tie. In this situation, simply play again until there is a clear winner.
