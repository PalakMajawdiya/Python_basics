```python
import random

def computer_guess(x):
  low = 1
  high = x
  feedback = ' '
  while feedback != 'c': # C for correct
    
    if low!=high:
      guess = random.randint(low, high)

    else:
      guess = low # could also bu high because low = high

    feedback = input(f'Is {guess} too high(H) , too low(L) or Correct(C) ???'.lower())

    if feedback == 'h':
      high = guess - 1 
      
    elif feedback == 'l':
      low = guess + 1

  print(f'Yay , the computer guess your number, {guess} Correctly!! ')

computer_guess(20)

```
