Hangman game is a great help in stimulating in learning foreign language particularly in teaching of vocabulary. This involves guessing letters to reveal a word or phrase. After a certain number of incorrect guesses, a character onscreen is "hanged".One player, the "host," makes up a secret word, while the other player tries to guess the word by asking what letters it contains. However, every wrong guess brings them one step closer to losing.
 Whenever the players guess a letter that is not in the secret word they get a strike that brings them closer to losing. To show this, the host draws a simple stick figure of a man being hung, adding a new part to the drawing with every wrong answer.If the players get every letter of the word before the host finishes drawing then they win. 

```python
import random
import string

words = ["aback","abaft","abandoned","abashed","aberrant","abhorrent","abiding","abject","ablaze","able","abnormal","aboard","aboriginal","abortive","abounding","abrasive","abrupt","absent","absorbed","absorbing","bad","barbarous","bashful","bawdy","beautiful","be-fitting","belligerent","beneficial","bent","berserk","best","better","bewildered","big","billowy","bite-sized","bitter","bizarre","black","cagey","calculating","callous","calm","capable","capricious","careful","careless","caring","cautious","ceaseless","certain","changeable","charming","cheap","cheerful","chemical","chief","daffy","daily","damaged","damaging","damp","dangerous","dapper","dark","dashing","dazzling","dead","deadpan","deafening","dear","debonair","decisive","decorous","eager","early","earsplitting","earthy","easy","eatable","economic","educated","efficacious","efficient","eight","elastic","elated","elderly","electric","elegant","elfin","fabulous","faded","faint","fair","faithful","fallacious","false","familiar","famous","fanatical","fancy","fantastic","far","far-flung","fascinated","fast","fat","faulty","gabby","gainful","gamy","gaping","garrulous","gaudy","general","gentle","giant","giddy","gifted","gigantic","glamorous","gleaming","glib","glistening","glorious","glossy","hushed","husky","hypnotic","hysterical","icky","icy","idiotic","ignorant","ill","illegal","ill-fated","ill-informed","illustrious"]



def get_valid_word(words):
  word = random.choice(words)  #it randomly chooses a word from the list

  while '-' in word or ' ' in word:
    word = random.choice(words)
  return word

def hangman():
  word = get_valid_word(words)
  word_letters = set(word) #keep track of guessed letters in the world
  alphabet = set(string.ascii_uppercase)
  used_letters = set() #what the user has guessed

  lives = 6

  #getting input from user 

  while len(word_letters) > 0 and lives > 0:

    #letters used
    #' '.join(['a','b','cd']) = 'a b cd'
    print('you have',lives,'lives left and you have used this letters: ',' '.join(used_letters))

    #what current word is that W _ R D
    word_list = [letter if letter in used_letters else '-' for letter in word]
    print('Current word: ',' '.join(word_list))

    user_letter = input('Guess a letter: ').upper()

    if user_letter in alphabet - used_letters: # if this is a valid character in the alphabet that i haven't used yet
      used_letters.add(user_letter) 
    
      if user_letter in  word_letters:
        word_letters.remove(user_letter)

      else:
        lives = lives - 1 #takes away a life if wrong
        print('Letter is not in word.')

    elif user_letter in used_letters:
     print('You have already used that character. Please try again')

    else:
      print("Invalid character.Please try again")

  if lives == 0:
      print('you died, sorry.The word was',word)
  else:
    print('you guessed the word',word,'!!')

hangman()
```
