Binary Search is a divide and conquer algorithm which actualy helps you search an ordered list faster by just scanning the element.For Binary search the list must be in sorted order

```python
def naive_search(l , target):
  for i in range(len(l)):
    if l[i] == target:
      return i

  return -1

if __name__=='__main__':
  l = [1,3,5,10,12]
  target = 10
  print(naive_search(l , target))
```

```python
import random
import time

def binary_search(l , target , low = None , high = None):
  if low is None:
    low = 0
  if high is None:
    high = len(l) - 1 

  if high<low:
    return -1

  midpoint = (low + high) // 2 # index of midpoint

  if l[midpoint] == target:
    return midpoint

  elif target < l[midpoint]:
    return binary_search(l , target , low , midpoint-1)

  else:
    return binary_search(l , target , midpoint+1 , high)
  

if __name__=='__main__':
  l = [1,3,5,10,12]
  target = 10

  print(binary_search(l , target))

  # buid a sorted list of length = 10000
  length = 10000
  sorted_list = set()
  while len(sorted_list) < length:
    sorted_list.add(random.randint(-3*length , 3*length))
  sorted_list = sorted(list(sorted_list))

  start = time.time()
  for target in sorted_list:
    naive_search(sorted_list , target)
  end = time.time()
  print("Naive search time is: ",(end-start)/length , 'seconds')


  start = time.time()
  for target in sorted_list:
    binary_search(sorted_list , target)
  end = time.time()
  print("Binary search time is: ",(end-start)/length , 'seconds')
```
